### Tesla Stock Price v.s. Search Volume

**Challenge:** Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes. Label one axis 'TSLA Stock Price' and the other 'Search Trend'. 

```python
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(
    df_tesla['MONTH'], 
    df_tesla['TSLA_WEB_SEARCH'], 
    color=mcolors.CSS4_COLORS['darkorange'], 
    label='Web Search'
)
ax2.plot(
    df_tesla['MONTH'], 
    df_tesla['TSLA_USD_CLOSE'], 
    color=mcolors.CSS4_COLORS['green'], 
    label='Stock Price'
)

ax1.set_ylabel('Number of Web Search')
ax2.set_ylabel('Stock Price')

plt.title('Tesla Web Search Trends vs. Stock Price')

# Combine legends from both axes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
rolled_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()
rolled_df.dropna(inplace=True)

plt.figure(figsize=(14, 8), dpi=120)
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate (R-6)', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.set_xlim((rolled_df.index.min(), rolled_df.index.max()))
ax1.grid(color='grey', linestyle='--')

ax1.set_ylabel('FRED U/E Rate', fontsize=14, color=mcolors.CSS4_COLORS['green'])
ax2.set_ylabel('Web Searches', fontsize=14, color=mcolors.CSS4_COLORS['darkorange'])
ax1.tick_params(labelsize=14)
ax2.tick_params(labelsize=14)

ax1.plot(
    rolled_df.index,
    rolled_df['UNRATE'],
    color=mcolors.CSS4_COLORS['green'],
    linestyle='dashdot',
    linewidth=3,
)
ax2.plot(
    rolled_df.index,
    rolled_df['UE_BENEFITS_WEB_SEARCH'],
    color=mcolors.CSS4_COLORS['darkorange'],
    linewidth=3,
)

plt.show()
```
