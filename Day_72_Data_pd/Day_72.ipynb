```python
clean_df = df.dropna()
clean_df.isna().sum()
```

```python
max_id_start =clean_df['Starting Median Salary'].idxmax()
clean_df.iloc[max_id_start]
```

```python
max_id_mid =clean_df['Mid-Career Median Salary'].idxmax()
clean_df.iloc[max_id_mid]
```

```python
min_id_start =clean_df['Starting Median Salary'].idxmin()
clean_df.iloc[min_id_start]
```

```python
min_id_mid =clean_df['Mid-Career Median Salary'].idxmin()
clean_df.iloc[min_id_mid]
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
high_risk = clean_df.sort_values('Spread', ascending=False)
high_risk[['Undergraduate Major', 'Spread']].head()
```

```python
high_pay = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_pay[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
group_by = clean_df.drop(columns='Undergraduate Major')
group_df = group_by.groupby('Group').mean()
pd.options.display.float_format = '{:,.2f}'.format
group_df
```
